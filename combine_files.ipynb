```python
import numpy 
import os
from astropy.io import fits
from astropy.table import Table
import astropy
import glob
```

```python
halo_dir = '/global/cscratch1/sd/chto100/BCC/'
halo_names = glob.glob(os.path.join(halo_dir,'Chinchilla-3_halos.??.fits')) + glob.glob(os.path.join(halo_dir,'Chinchilla-3_halos.?.fits'))
```

```python
example_file = fits.open(halo_names[0])
```

```python
example_table = Table(example_file[1].data)
```

```python
example_table
```

```python
halo_table_list = [Table(fits.open(halo_name)[1].data) for halo_name in halo_names]
```

```python
halo_table_combined = astropy.table.vstack(halo_table_list)
```

```python
lamb_table = Table(fits.open('./buzzard-1.9.8_3y3a_run_halos_lambda_chisq.fit')[1].data)
profile_table = Table(fits.open('./buzzard-1.9.8_3y3a_run_halos_profiles.fit')[1].data)
```

```python
lamb_profile_table = astropy.table.join(lamb_table, profile_table, keys_left='HALOID', keys_right='haloid')
```

```python
lamb_profile_table
```

```python
joint_lamb_profile_table = astropy.table.join(lamb_profile_table, halo_table_combined, keys_left='HALOID', keys_right='ID')
```

```python
joint_lamb_profile_table
```

```python
joint_lamb_profile_table.write('./combined_buzzard.fits', overwrite=True, format='fits')
```
